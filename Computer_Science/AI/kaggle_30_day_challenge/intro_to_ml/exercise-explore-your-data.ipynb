```python
home_data['Quarter'] = home_data['MoSold'].apply(lambda x: (x-1)//3 + 1)
df = home_data.groupby(['YrSold', 'Quarter']).size()\
                        .reset_index()\
                        .rename(columns={0:'NumSold'})
df.head()
```

```python
sns.catplot(data=df, x='YrSold', y='NumSold', hue='Quarter', kind='bar')
```

## Is the data helpful

```python
## Taking the sales avarage in every month of the year
avg_price_by_year_mo = pd.DataFrame(home_data.groupby(["YrSold", "MoSold"]).SalePrice.mean().round())
## Taking the counts of sales in every month of the year
cnt_price_by_year_mo = pd.DataFrame(home_data.groupby(["YrSold", "MoSold"]).SalePrice.count().round())
```

```python
## Displaying in one table
_table = avg_price_by_year_mo.merge(cnt_price_by_year_mo, on=["YrSold", "MoSold"])
_table = _table.rename(index = str, columns = {"SalePrice_x":"AVG_Price", "SalePrice_y":"SaleCount"})
_table
```

```python
## plotting every years month wise sales count
new_data = pd.DataFrame(columns=["YrMo", "Counts"])
## it is a multi index data

#sns.displot(_table.unstack(level=0)['AVG_Price'])
x = _table.unstack(level=0)['SaleCount']
for i in range(2006, 2011):
    for j in range(1, 13):
        string = f"{i}-{j}"
        new_data.loc[len(new_data.index)] = [string, x[str(i)][str(j)]]
new_data.head()
```

```python
## visualising the data
fig = plt.figure(figsize=(100,50))
plt.bar(new_data['YrMo'], new_data['Counts'], width=0.8, align='center')
plt.xticks(fontsize=40, rotation=45)
plt.yticks(fontsize=40)
plt.show()
```

```python
# Path of the file to read
iowa_file_path = '../input/home-data-for-ml-course/train.csv'

# Fill in the line below to read the file into a variable home_data
homedata = pd.read_csv(iowa_file_path)

Avgpricebyyearmonth = pd.DataFrame(homedata.groupby(["YrSold","MoSold"]).SalePrice.mean().round()) 
Cntpricebyyearmonth = pd.DataFrame(homedata.groupby(["YrSold","MoSold"]).SalePrice.count().round())

pricetable = Avgpricebyyearmonth.merge(Cntpricebyyearmonth,on = ["YrSold","MoSold"]) 
pricetable = pricetable.rename(index = str, columns = {"SalePrice_x": "AvgSalePrice", "SalePrice_y": "SaleCount" })

pricetable.reset_index(inplace=True) ### line added for simplicity

pricetable['Period'] = pd.to_datetime(pricetable['YrSold'].astype(str) + pricetable['MoSold'].astype(str),
format='%Y%m', errors='ignore')
pricetable['Period'] = pricetable['Period'].dt.strftime("%Y-%m")


import matplotlib.pyplot as plt

pricetable.plot.bar(x="Period",y="SaleCount",legend=None, figsize=(20, 10), color='navy')
plt.xticks(rotation=45)
plt.title("SaleCount by Year and Month", fontsize = 20)
plt.show()
```
